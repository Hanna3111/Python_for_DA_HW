3.1. Створіть зводну таблицю, яка покаже, скільки покупок (інвойсів) було зроблено в кожній з філій (`Branch`) за різними категоріями товарів. Запишіть таблицю в змінну `invoices_by_category` і виведіть змінну на екран.
Ця таблиця допоможе проаналізувати, в якій філії купують найбільше товарів кожної з категорій.

```python
invoices_by_category = df.pivot_table(index='Branch', columns='Product line', values='Invoice ID', aggfunc='count')
invoices_by_category
```

**Питання2:** Який платіжний метод найбільш популярний серед покупців у кожному місті?

**Гіпотеза2:** У великих містах люди частіше користуються безготівковими методами оплати (наприклад, картками чи електронними гаманцями), тоді як у менших містах популярніша готівка.

**Виконання:** Для цього ми можемо скористатись groupby() для групування за містом та платіжним методом і порахувати кількість використань кожного платіжного методу.

```python
payment_method_by_city = df.groupby(['City', 'Payment']).size()
payment_method_by_city
```

```python
most_popular_payment_method = payment_method_by_city.groupby('City').idxmax()
most_popular_payment_method
```

**Рекомендації:** Знати популярні платіжні методи дозволить магазинам налаштувати зручність оплати, пропонуючи найбільш затребувані способи. Якщо в одному місті популярні картки, можна зробити акцент на покращенні платіжних терміналів, а якщо в іншому — на готівкових розрахунках.

**Питання3:** Яка категорія товарів приносить найбільший прибуток в кожному філіалі?

**Гіпотеза3:** У кожному філіалі найбільший прибуток генерується однією категорією товарів. Наприклад, в одному філіалі це можуть бути "Електронні аксесуари", в іншому – "Продукти харчування".

**Виконання:** потрібно підсумувати прибуток (gross income) для кожної категорії товарів по кожному філіалу.

```python
grouped = df.groupby(['Branch', 'Product line'])['gross income'].sum()
grouped
```

```python
max_profit_category = grouped.groupby('Branch').idxmax()
max_profit_category
```

```python
max_profit_values = grouped[max_profit_category]
max_profit_values
```
